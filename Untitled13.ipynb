```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import random

# --- 1. Generate Synthetic Customer Data ---
# For demonstration purposes, we'll create a synthetic dataset
# that resembles customer purchase history, focusing on 'Annual Income (k$)'
# and 'Spending Score (1-100)'.
print("Generating synthetic customer data...")
np.random.seed(42) # for reproducibility

num_customers = 200
data = {
    'CustomerID': range(1, num_customers + 1),
    'Gender': [random.choice(['Male', 'Female']) for _ in range(num_customers)],
    'Age': np.random.randint(18, 70, num_customers),
    'Annual Income (k$)': np.random.randint(15, 150, num_customers),
    'Spending Score (1-100)': np.random.randint(1, 100, num_customers)
}
df = pd.DataFrame(data)

print("Synthetic data generated. First 5 rows:")
print(df.head())
print("\nData description:")
print(df.describe())

# --- 2. Select Features for Clustering ---
# We'll use 'Annual Income (k$)' and 'Spending Score (1-100)' for clustering.
# These columns are often used in customer segmentation examples.
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

print(f"\nFeatures selected for clustering: {X.columns.tolist()}")

# --- 3. Preprocessing: Scaling the Data ---
# It's crucial to scale the data before applying K-means,
# especially when features have different scales, to prevent
# features with larger values from dominating the distance calculations.
print("Scaling the data...")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

print("Scaled data first 5 rows:")
print(X_scaled_df.head())

# --- 4. Determine Optimal Number of Clusters (Elbow Method) ---
# The Elbow Method is a common technique to find the optimal number of clusters (K).
# It involves plotting the Within-Cluster Sum of Squares (WCSS) against the
# number of clusters and looking for an "elbow" point where the rate of
# decrease in WCSS significantly changes.
print("\nDetermining optimal number of clusters using the Elbow Method...")
wcss = []
# Test K from 1 to 10
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10) # n_init to suppress warning
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_) # inertia_ is the WCSS

# Plotting the Elbow Method graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.show()

print("Please observe the plot to identify the 'elbow' point.")
print("A common choice for this type of data is often 3, 4, or 5 clusters.")

# --- 5. Apply K-Means Clustering ---
# Based on the elbow method (let's assume K=5 for this example,
# but you can adjust based on your observation of the plot).
optimal_k = 5 # This should be chosen based on the Elbow Method plot
print(f"\nApplying K-Means clustering with K = {optimal_k}...")

kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42, n_init=10)
# Fit K-means to the scaled data and predict the clusters
clusters = kmeans.fit_predict(X_scaled)

# Add the cluster labels back to the original DataFrame
df['Cluster'] = clusters

print("Clustering complete. First 5 rows with cluster assignments:")
print(df.head())
print("\nDistribution of customers per cluster:")
print(df['Cluster'].value_counts().sort_index())

# --- 6. Visualize the Clusters ---
# Plotting the clusters to see the segmentation visually.
# We'll use the original (unscaled) data for better interpretability on the plot axes.
print("\nVisualizing the clusters...")
plt.figure(figsize=(12, 8))
# Iterate through each cluster and plot its points
for i in range(optimal_k):
    plt.scatter(df[df['Cluster'] == i]['Annual Income (k$)'],
                df[df['Cluster'] == i]['Spending Score (1-100)'],
                label=f'Cluster {i}', s=100, alpha=0.7)

# Plot the cluster centroids (transformed back to original scale for plotting)
# Inverse transform the centroids from scaled back to original scale for plotting
centroids_scaled = kmeans.cluster_centers_
centroids_original_scale = scaler.inverse_transform(centroids_scaled)

plt.scatter(centroids_original_scale[:, 0], centroids_original_scale[:, 1],
            s=300, c='red', marker='X', label='Centroids', edgecolor='black')

plt.title(f'Customer Segments (K-Means Clustering with K={optimal_k})')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid(True)
plt.show()

print("\nK-Means clustering successfully applied and visualized!")
print("Each color represents a different customer segment.")
print("The red 'X' marks are the centroids (average points) of each cluster.")

# --- 7. Analyze Cluster Characteristics (Optional) ---
# You can further analyze the characteristics of each cluster
# to understand the customer segments.
print("\nAnalyzing characteristics of each cluster:")
for i in range(optimal_k):
    print(f"\n--- Cluster {i} ---")
    cluster_data = df[df['Cluster'] == i]
    print(cluster_data[['Annual Income (k$)', 'Spending Score (1-100)', 'Age']].describe())
    print(f"Gender distribution in Cluster {i}:")
    print(cluster_data['Gender'].value_counts(normalize=True))

```
